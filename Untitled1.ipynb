```python
import matplotlib.pyplot as plt

x = [4, 5, 6, 7, 8, 7, 14]
y = [15, 21, 24, 17, 25, 26, 21]

plt.scatter(x[:7], y)  # Make sure both lists have the same number of elements
plt.show()

```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = list(zip(x, y))
iner = []

for i in range(1, 8):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    iner.append(kmeans.inertia_)

plt.plot(range(1, 8), iner, marker='o')
plt.title("Elbow method")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```
